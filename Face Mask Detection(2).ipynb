```python
import numpy as np
import cv2
```

```python
with_mask = np.load('with_mask.npy')
without_mask = np.load('without_mask.npy')
```

```python
# Converting 4-D image into 2-D
```

```python
with_mask = with_mask.reshape(200,50*50*3)
without_mask = without_mask.reshape(200,50*50*3)
```

```python
with_mask.shape
```

```python
without_mask.shape
```

```python
x = np.r_[with_mask, without_mask]                   # r_ is used to concatenate the rows with_mask, without_mask
```

```python
x.shape            # Training data is ready
```

```python
labels = np.zeros(x.shape[0])   # First 200 will be zero with_mask
```

```python
labels[200:] = 1.0             # Next 200 will be one bcz  without_mask
```

```python
names = {0 : 'Mask', 1 : 'No Mask'}       # If o/p is 0 then its face with mask & viceversa
```

```python
from sklearn.svm import SVC              
from sklearn.metrics import accuracy_score     
```

```python
from sklearn.model_selection import train_test_split   # divide data into training & testing
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, labels, test_size = 0.25)
```

```python
x_train.shape  
```

```python
# bcz of 7500 coloums ( as in o/p) , speed of machine will be reduced
#     So to overcome this , we will use Dimentionality Reductions (To reduce the data )
#                For this ,   Import Package Decomposition
```

```python
from sklearn.decomposition import PCA            
#Principal Component Analysis
```

```python
pca = PCA(n_components=3)
x_train = pca.fit_transform(x_train)
```

```python
svm = SVC()
svm.fit(x_train, y_train)
```

```python
haar_data = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
capture = cv2.VideoCapture(0)
data = []
while True:
    flag, img = capture.read()
    if flag:
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img, (x,y),(x+w, y+h),(255,0,255), 4)
            face = img[y:y+h, x:x+w, :]
            face = cv2.resize(face, (50,50))
            face = face.reshape(1,-1)
            #face = pca.transform(face)
            pred = svm.predict(face)[0]
            n = names[int(pred)]
            if n == "Mask":
                cv2.putText(img,"Mask",(x,y-30),cv2.FONT_HERSHEY_SIMPLEX,2,(0,0,255),3)
                print(n)
            else:
                cv2.putText(img,"No Mask",(x,y-30),cv2.FONT_HERSHEY_SIMPLEX,2,(0,0,255),3)
                print(n)
        cv2.imshow('result', img)
        # 27 - ASCII of Escape
        if cv2.waitKey(2) == 27:
            break
            
capture.release()
cv2.destroyAllWindows()
```
